```python
import pandas as pd
from pathlib import Path
from typing import Union, Dict, Any

def load_healthcare_diseases() -> Union[Dict[str, Any], str]:
    """
    Loads a comprehensive healthcare dataset from HuggingFace with
    treatment, symptoms, risk factors, complications, prevention, etc.
    """
    # Use current working directory instead of __file__
    file_path = Path.cwd() / "healthdatasets" / "healthcare_diseases_dataset_huggingface.csv"
    if not file_path.exists():
        return f"Healthcare dataset not found at {file_path}"

    data = pd.read_csv(file_path)

    # Basic validation
    if "disease" not in data.columns:
        return "Healthcare dataset missing required 'disease' column"
    
    data_dict = data.to_dict(orient="index")
    return data_dict

data = load_healthcare_diseases()

# Since data could be a string (error), check before printing columns
if isinstance(data, dict):
    print(list(next(iter(data.values())).keys()))
else:
    print(data)
```

```python
def load_health_dataset() -> Union[Dict[str, Any], str]:
    """
    Loads worldwide disease dataset with symptoms (columns: disease, symptoms 1-12)
    Uses caching to prevent repeated file reads.
    """
    file_path = Path.cwd()/"healthdatasets"/"health_dataset.csv"
    if not file_path.exists():
        return f"Health dataset not found at {file_path}"
    
    data = pd.read_csv(file_path, engine="python")
    data_dict = data.to_dict(orient="index")

    return data_dict

data = load_health_dataset()
if isinstance(data, dict):
    print(list(next(iter(data.values())).keys()))
else:
    print(data)
```
